```python
import pandas as pd
import requests
import json

#gathers a creates dataframe for practitioners dataset

headers = {
    "Accept": "application/json",
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}
           
url = f'https://dw.euro.who.int/api/v3/measures/HLTHRES_67?output=data'

url_text = requests.get(url, headers=headers).text

result = json.loads(url_text)
#print(result)

data = result['data']

#print(data)
df_practitioners = pd.DataFrame()
final_dict = dict()

for item in data:
    
    final_dict['country'] = item['dimensions']['COUNTRY']
    #final_dict['country_grp'] = item['dimensions']['COUNTRY_GRP']
    final_dict['year'] = item['dimensions']['YEAR']
    final_dict['practitioners'] = item['value']['numeric']
    series = pd.Series(final_dict)
    df_practitioners = pd.concat([df_practitioners, series.to_frame().T], ignore_index = True)
    
    

#print(series)
#pd.set_option('display.max_rows', None)
df_practitioners.tail(8)


```

```python
import plotly.express as px
df_practitioners2 = df_practitioners.iloc[166:]
figure = px.scatter(x=df_practitioners2['year'], y=df_practitioners2['practitioners'], color=df_practitioners2['country'], labels={
                     "x": "Year",
                     "y": "Number of Practitioners"
                 },title = 'Generalist medical practitioners, per 10,000 population')
figure.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(25, 12))
seen_country = []
for item in df_practitioners2['country']:
    if item not in seen_country:
        subset = df_practitioners2[(df_practitioners2['country'] == item)].copy()
        plt.plot(subset['year'], subset['practitioners'], label=item)
        seen_country.append(item)
plt.xlabel('Year')
plt.ylabel('Number of Practitioners')
plt.title('Generalist medical practitioners, per 10,000 population')
plt.legend()
plt.show()
```

```python
import pandas as pd
import requests
import json
#gathers a creates dataframe for expenditure dataset
headers = {
    "Accept": "application/json",
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}
           
url2 = f'https://dw.euro.who.int/api/v3/measures/HFA_570?output=data'

url_text2 = requests.get(url2, headers=headers).text

result2 = json.loads(url_text2)
#print(result2)

data = result2['data']

#print(data)
df_expenditure = pd.DataFrame()
final_dict = dict()

for item in data:
    
    final_dict['country'] = item['dimensions']['COUNTRY']
    #final_dict['country_grp'] = item['dimensions']['COUNTRY_GRP']
    final_dict['year'] = item['dimensions']['YEAR']
    #final_dict['sex'] = item['dimensions']['SEX']
    final_dict['expenditure'] = item['value']['numeric']
    series = pd.Series(final_dict)
    df_expenditure = pd.concat([df_expenditure, series.to_frame().T], ignore_index = True)
    
    

#print(series)
#pd.set_option('display.max_rows', None)
df_expenditure.head()



```

```python
import pandas as pd
import requests
import json
#gathers a creates dataframe for births dataset
headers = {
    "Accept": "application/json",
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}
           
url = f'https://dw.euro.who.int/api/v3/measures/HFA_16?output=data'

url_text = requests.get(url, headers=headers).text

result = json.loads(url_text)
#print(result2)

data = result['data']

#print(data)
df_births = pd.DataFrame()
final_dict = dict()

for item in data:
    
    final_dict['country'] = item['dimensions']['COUNTRY']
    #final_dict['country_grp'] = item['dimensions']['COUNTRY_GRP']
    final_dict['year'] = item['dimensions']['YEAR']
    final_dict['births'] = item['value']['numeric']
    series = pd.Series(final_dict)
    df_births = pd.concat([df_births, series.to_frame().T], ignore_index = True)
    
    

#print(series)
#pd.set_option('display.max_rows', None)
df_births.head()


#births per 1000 population
```

```python
import pandas as pd
import requests
import json
#gathers a creates dataframe for catastrophic spending dataset
headers = {
    "Accept": "application/json",
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}
           
url = f'https://dw.euro.who.int/api/v3/measures/UHCFP_4?output=data'

url_text = requests.get(url, headers=headers).text

result = json.loads(url_text)
#print(result2)

data = result['data']

#print(data)
df_catastrophic = pd.DataFrame()
final_dict = dict()

for item in data:
    
    final_dict['country'] = item['dimensions']['COUNTRY']
    #final_dict['country_grp'] = item['dimensions']['COUNTRY_GRP']
    final_dict['year'] = item['dimensions']['YEAR']
    final_dict['catastrophic spending'] = item['value']['numeric']
    series = pd.Series(final_dict)
    df_catastrophic = pd.concat([df_catastrophic, series.to_frame().T], ignore_index = True)
    df_catastrophic = df_catastrophic.sort_values('year', ascending=True)
    
    

#print(series)
#pd.set_option('display.max_rows', None)
df_catastrophic.head()


```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(30, 12))
seen_country = []
for item in df_catastrophic['country']:
    if item not in seen_country:
        subset = df_catastrophic[(df_catastrophic['country'] == item)].copy()
        plt.plot(subset['year'], subset['catastrophic spending'], label=item)
        seen_country.append(item)
plt.xlabel('Year')
plt.ylabel('Catastrophic Spending')
plt.title('Catastrophic health spending, all households')
plt.legend()
plt.show()
```
